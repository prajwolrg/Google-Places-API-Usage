```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
def genData():
    x=[]
    y=[]
    r = 3

    x.append(0)
    y.append(0)

    x.append(r+r*math.cos(math.pi/3))
    y.append(-r * math.cos(math.pi/6))

    x.append(0)
    y.append(-2*r*math.cos(math.pi/6))

    x.append(-r -r*math.cos(math.pi/3))
    y.append(-r * math.cos(math.pi/6))

    x.append(-r -r*math.cos(math.pi/3))
    y.append(r*math.cos(math.pi/6))

    x.append(0)
    y.append(2*r*math.cos(math.pi/6))

    x.append(r+r*math.cos(math.pi/3))
    y.append(r * math.cos(math.pi/6))

    x.append(2*r+2*r*math.cos(math.pi/3))
    y.append(0)

    x.append(2*r+2*r*math.cos(math.pi/3))
    y.append(-2*r*math.cos(math.pi/6))

    x.append(r+r*math.cos(math.pi/3))
    y.append(-r-2*r * math.cos(math.pi/6))

    return({'x':x, 'y':y})
```

```python
def genDataAlt():
    x=[]
    y=[]
    r = 3

    largestep = 3*r/2
    smallstep = math.sqrt(3)*r/2
    
    xvalue = 0
    yvalue = 0

    x.append(xvalue)
    y.append(yvalue)

    xvalue += largestep
    yvalue -= smallstep
    x.append(xvalue)
    y.append(yvalue)

    for i in range(1,4):
        for j in range(i):
            xvalue -=largestep
            yvalue -= smallstep
            x.append(xvalue)
            y.append(yvalue)

        for j in range(i):
            xvalue -= largestep
            yvalue += smallstep
            x.append(xvalue)
            y.append(yvalue)
       
        for j in range(i):
            yvalue += 2*smallstep
            x.append(xvalue)
            y.append(yvalue)

        for j in range(i):
            xvalue +=largestep
            yvalue += smallstep
            x.append(xvalue)
            y.append(yvalue)

        for j in range(i):
            xvalue +=largestep
            yvalue -= smallstep
            x.append(xvalue)
            y.append(yvalue)

        for j in range(i):
            xvalue +=largestep
            yvalue -= smallstep
            x.append(xvalue)
            y.append(yvalue)

        for j in range(i):
            yvalue -= 2*smallstep
            x.append(xvalue)
            y.append(yvalue)

    # for i in range(2):


    return({'x':x, 'y':y})
```

```python
points = genDataAlt()
fig, ax = plt.subplots() # note we must use plt.subplots, not plt.subplot
SIZE = 40
for i in range(len(points['x'])):
    circle = plt.Circle((points['x'][i], points['y'][i]), 3, color = 'b', alpha=0.2)
    ax.add_artist(circle)

plt.xlim(-SIZE, SIZE)
plt.ylim(-SIZE, SIZE)
plt.gca().set_aspect('equal', adjustable='box')


plt.show()
```
